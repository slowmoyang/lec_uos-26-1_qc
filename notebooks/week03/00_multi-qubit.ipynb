# Week 3: Multi-qubit Gate Examples

This notebook demonstrates S gate, T gate, CNOT gate, Bell-state generation, and Toffoli gate using Qiskit.

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import matplotlib.pyplot as plt
```

## 1. $S$-gate

```python
# Create a 1-qubit circuit
qc_s = QuantumCircuit(1, 1) # N_qubit=1, M_cbit=1

# Prepare |+> = (|0> + |1>) / sqrt(2)
qc_s.h(0)

# Apply S gate
qc_s.s(0)

# Measure
qc_s.measure(0, 0)

qc_s.draw("mpl")
```

```python
# Statevector before measurement
qc_s_state = QuantumCircuit(1)

state_s = Statevector.from_instruction(qc_s_state)
print("Statevector for S gate example:")
print(state_s)

plot_bloch_multivector(state_s)
```

```python
qc_s_state.h(0)

state_s = Statevector.from_instruction(qc_s_state)
print("Statevector for S gate example:")
print(state_s)

plot_bloch_multivector(state_s)
```

```python
qc_s_state.s(0)

state_s = Statevector.from_instruction(qc_s_state)
print("Statevector for S gate example:")
print(state_s)

plot_bloch_multivector(state_s)
```

```python
# Sampling
sampler = StatevectorSampler()
result_s = sampler.run([qc_s], shots=1000).result()

counts_s = result_s[0].data.c.get_counts()
print("Measurement counts:", counts_s)

plot_histogram(counts_s)
```

## 2. $T$ gate

```python
# Create a 1-qubit circuit
qc_t = QuantumCircuit(1, 1)

# Prepare |+>
qc_t.h(0)

# Apply T gate
qc_t.t(0)

# Measure
qc_t.measure(0, 0)

qc_t.draw("mpl")
```

```python
# Statevector before measurement
qc_t_state = QuantumCircuit(1)
state_t = Statevector.from_instruction(qc_t_state)
plot_bloch_multivector(state_t)
```

```python
qc_t_state.h(0)
state_t = Statevector.from_instruction(qc_t_state)
plot_bloch_multivector(state_t)
```

```python
qc_t_state.t(0)
state_t = Statevector.from_instruction(qc_t_state)
plot_bloch_multivector(state_t)
```

```python
print("Statevector for T gate example:")
print(state_t)
```

```python
# Sampling
result_t = sampler.run([qc_t], shots=1000).result()

counts_t = result_t[0].data.c.get_counts()
print("Measurement counts:", counts_t)

plot_histogram(counts_t)
```

## 3. CNOT gate example

flips target when control is 1

```python
# Create a 2-qubit circuit
qc_cnot = QuantumCircuit(2, 2)

# Put qubit 0 in |1>
qc_cnot.x(0)

# Apply CNOT: control=0, target=1
qc_cnot.cx(0, 1)

# Measure both qubits
qc_cnot.measure([0, 1], [0, 1])

qc_cnot.draw("mpl")
```

```python
# Statevector before measurement
qc_cnot_state = QuantumCircuit(2)
qc_cnot_state.x(0)
qc_cnot_state.cx(0, 1)

state_cnot = Statevector.from_instruction(qc_cnot_state)
print("Statevector for CNOT example:")
print(state_cnot)
```

```python
# Sampling
result_cnot = sampler.run([qc_cnot], shots=1000).result()

counts_cnot = result_cnot[0].data.c.get_counts()
print("Measurement counts:", counts_cnot)

plot_histogram(counts_cnot)
```

## 4. Bell state generation circuit

creates an entangled 2-qubit state

```python
# Create a 2-qubit circuit
qc_bell = QuantumCircuit(2, 2)

# Bell state: (|00> + |11>) / sqrt(2)
qc_bell.h(0)
qc_bell.cx(0, 1)

# Measure both qubits
qc_bell.measure([0, 1], [0, 1])

print(qc_bell)
qc_bell.draw("mpl")
```

```python
# Statevector before measurement
qc_bell_state = QuantumCircuit(2)
qc_bell_state.h(0)
qc_bell_state.cx(0, 1)

state_bell = Statevector.from_instruction(qc_bell_state)
print("Statevector for Bell state:")
print(state_bell)
```

```python
# Sampling
result_bell = sampler.run([qc_bell], shots=1000).result()

counts_bell = result_bell[0].data.c.get_counts()
print("Measurement counts:", counts_bell)

plot_histogram(counts_bell)
```

## 5. Toffoli gate example

flips target when both controls are 1

```python
# Create a 3-qubit circuit
qc_ccx = QuantumCircuit(3, 3)

# Prepare control qubits in |1>, |1>
qc_ccx.x(0)
qc_ccx.x(1)

# Apply Toffoli gate: controls=0,1 target=2
qc_ccx.ccx(0, 1, 2)

# Measure all qubits
qc_ccx.measure([0, 1, 2], [0, 1, 2])

qc_ccx.draw("mpl")
```

```python
# Statevector before measurement
qc_ccx_state = QuantumCircuit(3)
qc_ccx_state.x(0)
qc_ccx_state.x(1)
qc_ccx_state.ccx(0, 1, 2)

state_ccx = Statevector.from_instruction(qc_ccx_state)
print("Statevector for Toffoli gate example:")
print(state_ccx)
```

```python
# Sampling
result_ccx = sampler.run([qc_ccx], shots=1000).result()

counts_ccx = result_ccx[0].data.c.get_counts()
print("Measurement counts:", counts_ccx)

plot_histogram(counts_ccx)
```
