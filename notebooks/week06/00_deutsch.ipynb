# Deutsch's algorithm

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
def oracle_1(qc):
    """
    Oracle for a constant function: f(x) = 0.
    This oracle leaves the target qubit unchanged.
    """
    pass

def oracle_2(qc):
    """
    Oracle for a balanced function: f(x) = x.
    If the input qubit is 1, flip the target qubit.
    """
    qc.cx(0, 1)

def oracle_3(qc):
    """
    Oracle for a balanced function: f(x) = x ⊕ 1.
    First apply f(x) = x, then flip the output once more.
    """
    qc.cx(0, 1)
    qc.x(1)

def oracle_4(qc):
    """
    Oracle for a constant function: f(x) = 1.
    Always flip the target qubit regardless of the input.
    """
    qc.x(1)
```

```python
oracle = oracle_4
```

# Build a circuit

```python
# Create a 2-qubit circuit with 1 classical bit for measurement

qc = QuantumCircuit(2, 1)

# Prepare the second qubit in |1>

qc.x(1)
qc.barrier()

# Apply Hadamard to the second qubit to create the |->
# state, which enables phase kickback
qc.h(1)
qc.barrier()

# Apply Hadamard to the first qubit to create a superposition
# of input states |0> and |1>
qc.h(0)
qc.barrier()

# Apply one of the Deutsch oracle circuits
oracle(qc)
qc.barrier()

# Apply Hadamard again to the first qubit to interfere amplitudes
# and reveal whether the function is constant or balanced
qc.h(0)
qc.barrier()

# Measure the first qubit
qc.measure(0, 0)
```

## Draw the circuit

```python
qc.draw('mpl')
```

## Run the circuit on the Aer simulator

```python
simulator = AerSimulator()

# Use one shot and return the measured bit string
job = simulator.run(qc, shots=1, memory=True)

result = job.result()

# Extract the measurement result
measurements = result.get_memory()
print(measurements)

# Interpret the result:
# 0 -> constant function
# 1 -> balanced function
print("constant" if measurements[0] == "0" else "balanced")
```
